```python
# Import Libraries
import os
import numpy as np
import pandas as pd 
```

```python

def fileSelect():
    try:
        # Select filename for analysis from working directory       
        quicktestFiles = []
        filenames = os.listdir()


        for filename in filenames:
            if filename.endswith('csv'):
                quicktestFiles.append(filename)
            elif filename.endswith('CSV'):
                quicktestFiles.append(filename)

        if (bool(quicktestFiles)) == False:
            print('No quicktest files to process - Goodbye!')
            exit()
        else:

            for n,each in enumerate(quicktestFiles):
                print(n,each)
            ask = int(input('Select quicktest CSV file for processing >'))
            quicktestFile = quicktestFiles[ask]
            print(quicktestFiles)
            #print('You selected:', quicktestFile)
            filename = quicktestFile
            return filename
    except IndexError:
        print('The number selected is invalid, try again')
        fileData(fileSelect())
    except ValueError:
        print('Invalid entry, try again')
        fileData(fileSelect())
    except NoneType:
        print('Invalid entry, try again')
        fileData(fileSelect())

def fileData(filename):

    # Strip file extionsion from filename
    furnaceInfo = filename[:-4].strip().upper().replace(' ', '-')

    # Split furnace info into furnace, run number and furnace run date
    furnaceInfo = furnaceInfo.split('-')
    furnace = furnaceInfo[0]
    if furnace == 'C':
        furnace = 'Rotovac'
    elif furnace == 'F1':
        furnace = 'Frank1'
    elif furnace == 'F2':
        furnace = 'Frank2'
    else:
        print('Bad filename')
        #  ????Send back to main loop????


    MO = str(input('Enter Manufacturing Order Number:'))
    runNo = furnaceInfo[1]
    month = furnaceInfo[2]
    day = furnaceInfo[3]
    year = furnaceInfo[4]
    date = (month + '/' + day + '/' + year)

    # Calculate mean and standard deviation of Resistance, Voltage, and Amps
    df = pd.read_csv(filename)  
    df[' Resistance'] = pd.to_numeric(df[' Resistance'], errors='coerce')
    df[' Volts'] = pd.to_numeric(df[' Volts'], errors='coerce')
    df[' Amps'] = pd.to_numeric(df[' Amps'], errors='coerce')
    df.replace('0', np.nan, inplace = True)
    df.dropna(inplace = True)
    amps = df[' Amps'].mean()
    std1 = df[' Amps'].std()
    volts = df[' Volts'].mean()
    res = df[' Resistance'].mean()
    std2 = df[' Resistance'].std()

    d = [MO]
    output = pd.DataFrame(d)
    output.columns = ['MO No']
    output['Furnace'] = furnace
    output['RunNo'] = runNo
    output['Run Date'] = date

    output['Mean Resistance'] = res
    output['Resistance Std'] = std2
    output['Volts'] = volts
    output['Mean Amps'] = amps
    output['Amps Std'] = std1

    #output.to_csv("quicktest_out.csv", mode = 'a', index=False, header=False)

    print(output)

        
```
